```python
data = np.random.rand(50, 3)
df = pd.DataFrame(data, columns=['A', 'B', 'C'])
num_nulls = int(0.1 * df.size)
null_indices = np.random.choice(df.size, num_nulls, replace=False)
df.values.ravel()[null_indices] = np.nan
missing_values = df.isnull().sum()
print("Missing values:\n", missing_values)
```

```python
row_sums = df.sum(axis=1)
max_sum_row_index = row_sums.idxmax()
df = df.drop(index=max_sum_row_index)
df
```

```python
df = df.sort_values(by='A')
print("\nDataFrame sorted by the first column (A):\n", df)
df = df.drop_duplicates(subset='A')
print("\nDataFrame after removing duplicates from the first column (A):\n", df)
```

```python
correlation = df['A'].corr(df['B'])
covariance = df['B'].cov(df['C'])
print(f"\nCorrelation between 'A' and 'B': {correlation}")
print(f"Covariance between 'B' and 'C': {covariance}")
df['B_discretized'] = pd.cut(df['B'], bins=5)
print("\nDataFrame with discretized 'B' column:\n", df)
```

```python
data1 = {'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'David', 'Bob'],
        'Date': ['2024-07-20', '2024-07-21', '2024-07-22', '2024-07-23', '2024-07-24', '2024-07-20'],
        'Duration': [30, 40, 50, 30, 40, 30]}
df1 = pd.DataFrame(data1)

data2 = {'Name': ['Bob', 'Charlie', 'Eve', 'Alice', 'Frank', 'Bob'],
        'Date': ['2024-07-20', '2024-07-21', '2024-07-22', '2024-07-23', '2024-07-24', '2024-07-25'],
        'Duration': [30, 40, 50, 30, 40, 50]}
df2 = pd.DataFrame(data2)

merged_df = pd.merge(df1, df2, on='Name', how='inner')
both_workshops = merged_df['Name'].unique()
print("Students who attended both workshops:\n", both_workshops)
```

```python

df1_only = df1[~df1['Name'].isin(both_workshops)]
df2_only = df2[~df2['Name'].isin(both_workshops)]
single_workshop = pd.concat([df1_only, df2_only])['Name'].unique()
print("\nStudents who attended only one workshop:\n", single_workshop)
```

```python
row_merged_df = pd.concat([df1, df2])
total_records = len(row_merged_df)
print("\nTotal number of records after row-wise merge:\n", total_records)
```

```python
row_merged_df = pd.concat([df1, df2])
row_merged_df = row_merged_df.set_index(['Name', 'Date'])
print("\nDescriptive statistics for hierarchical data frame:\n", row_merged_df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]

print(iris_df.info())
```

```python
print(iris_df.isnull().sum())
plt.figure(figsize=(8, 6))
iris_df['species'].value_counts().plot(kind='bar')
plt.title('Frequency of Iris Species')
plt.xlabel('Species')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(x='sepal length (cm)', y='petal length (cm)', data=iris_df)
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(iris_df['petal width (cm)'])
plt.title('Density Distribution of Petal Width')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.pairplot(iris_df, hue='species')
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.heatmap(iris_df[['sepal length (cm)', 'sepal width (cm)']].corr(), annot=True, cmap='viridis')
plt.title('Heatmap of Sepal Length and Sepal Width Correlation')
plt.show()
```

```python
numeric_features = iris_df.select_dtypes(include=['float64', 'int64'])
for col in numeric_features.columns:
    mean = numeric_features[col].mean()
    median = numeric_features[col].median()
    std_dev = numeric_features[col].std()
    mode = numeric_features[col].mode()[0]

    print(f"\nFeature: {col}")
    print(f"Mean: {mean:.2f}")
    print(f"Median: {median:.2f}")
    print(f"Mode: {mode:.2f}")
    print(f"Standard Deviation: {std_dev:.2f}")

    cm = sm.stats.DescrStatsW(numeric_features[col])
    ci_low, ci_high = cm.tconfint_mean()
    print(f"95% Confidence interval: ({ci_low:.2f}, {ci_high:.2f})")
    print(f"Standard Error: {cm.std_mean:.2f}")
```

```python

numerical_iris_df = iris_df.select_dtypes(include=['number'])

plt.figure(figsize=(10, 8))
sns.heatmap(numerical_iris_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Iris Features')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
titanic_df = pd.read_csv("/content/Titanic-Dataset.csv")
column_with_most_missing = titanic_df.isnull().sum().idxmax()
titanic_df = titanic_df.drop(columns=column_with_most_missing)
```

```python

passengers_over_30 = len(titanic_df[titanic_df['Age'] > 30])
print(f"Total passengers with age > 30: {passengers_over_30}")
```

```python
total_fare_second_class = titanic_df[titanic_df['Pclass'] == 2]['Fare'].sum()
print(f"Total fare paid by second-class passengers: {total_fare_second_class}")
survivors_by_class = titanic_df.groupby('Pclass')['Survived'].sum()
print("\nNumber of survivors per passenger class:")
print(survivors_by_class)
```

```python

age_stats_gender = titanic_df.groupby('Sex')['Age'].describe()
print("\nDescriptive statistics for age, gender-wise:\n", age_stats_gender)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Fare', y='Sex', data=titanic_df)
plt.title('Passenger Fare vs. Gender')
plt.xlabel('Fare')
plt.ylabel('Gender')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(titanic_df['Age'], label='Age')
sns.kdeplot(titanic_df['Fare'], label='Fare')
plt.title('Density Distribution of Age and Fare')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
class_counts = titanic_df['Pclass'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(class_counts, labels=[f'Class {i}' for i in class_counts.index], autopct='%1.1f%%', colors=['lightblue', 'lightcoral', 'lightgreen'])
plt.title('Distribution of Passenger Classes')
plt.show()
```

```python

survival_rate_by_class = titanic_df.groupby('Pclass')['Survived'].mean() * 100
print("\nSurvival rate by class:")
print(survival_rate_by_class)
print("\nDid class play a role in survival? Yes, as the survival rate varies significantly across passenger classes.")
```

```python
import pandas as pd

data = {'FamilyName': ['Shah', 'Vats', 'Vats', 'Kumar', 'Vats', 'Kumar', 'Shah', 'Shah', 'Kumar', 'Vats'],
        'Gender': ['Male', 'Male', 'Female', 'Female', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male'],
        'MonthlyIncome (Rs.)': [44000.00, 65000.00, 43150.00, 66500.00, 255000.00, 103000.00, 55000.00, 112400.00, 81030.00, 71900.00]}

df = pd.DataFrame(data)
familywise_income = df.groupby('FamilyName')['MonthlyIncome (Rs.)'].sum()
print("Familywise Gross Monthly Income:\n", familywise_income)
```

```python
def range_income(x):
    return f"{x.min()}-{x.max()}"

family_income_range = df.groupby('FamilyName')['MonthlyIncome (Rs.)'].agg(range_income)
print("\nIncome Range for Each Family:\n", family_income_range)
income_less_than_80000 = df[df['MonthlyIncome (Rs.)'] < 80000.00]
print("\nMonthly Income of Members Earning Less than Rs. 80000.00:\n", income_less_than_80000)
```

```python
female_stats = df[df['Gender'] == 'Female']['MonthlyIncome (Rs.)']
total_females = len(female_stats)
avg_female_income = female_stats.mean()
print(f"\nTotal Number of Females: {total_females}")
print(f"Average Monthly Income of Females: {avg_female_income}")
```

```python
average_income = df['MonthlyIncome (Rs.)'].mean()
df_filtered = df[df['MonthlyIncome (Rs.)'] >= average_income]
print("\nDataFrame after removing rows with income less than average income:\n", df_filtered)
```
