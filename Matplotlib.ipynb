```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')
plt.title("3D Surface Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)  # 1000 random numbers

plt.hist(data, bins=30, color='blue', alpha=0.7)
plt.title("Simple Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(data, bins=10, color='green', alpha=0.7, edgecolor='black')
plt.title("Histogram with 10 Bins")
plt.show()
```

```python
plt.hist(data, bins=20, density=True, color='purple', alpha=0.6)
plt.title("Density Histogram")
plt.show()
```

```python
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 2  # Shifted dataset

plt.hist([data1, data2], bins=30, color=['blue', 'red'], alpha=0.7, label=['Dataset 1', 'Dataset 2'])
plt.title("Overlapping Histograms")
plt.legend()
plt.show()
```

```python
plt.hist(data, bins=30, orientation='horizontal', color='cyan', alpha=0.6)
plt.title("Horizontal Histogram")
plt.show()
```

```python
weights = np.ones_like(data) / len(data)  # Normalize weights
plt.hist(data, bins=30, weights=weights, color='blue', alpha=0.7)
plt.title("Weighted Histogram")
plt.ylabel("Probability")
plt.show()
```

```python
plt.hist(data, bins=30, histtype='stepfilled', color='black', linewidth=1.5)
plt.title("Step Histogram")
plt.show()
```

```python
plt.hist(data, bins=20, histtype='bar', color='blue', alpha=0.5, edgecolor='black', label='Bar')
plt.hist(data, bins=20, histtype='stepfilled', color='red', alpha=0.5, label='Stepfilled')
plt.title("Overlay of Bar and Stepfilled Histograms")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 10000 samples
rand_data = np.random.rand(10000)
randn_data = np.random.randn(10000)

# Plot histograms
plt.figure(figsize=(12, 5))

# Uniform distribution (rand)
plt.subplot(1, 2, 1)
plt.hist(rand_data, bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.title("Uniform Distribution (rand)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Normal distribution (randn)
plt.subplot(1, 2, 2)
plt.hist(randn_data, bins=30, color='red', alpha=0.7, edgecolor='black')
plt.title("Normal Distribution (randn)")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
